## Importación de datos

```python
# Librerias
import pandas as pd
import numpy as np
from scipy.stats import kurtosis
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Datos
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json"

# DataFrame
df = pd.read_json(url)

# Revisión
df.head()
```

```python
# Columnas del dataset
df.columns
```

```python
# Columnas no normalizadas
normal = df.columns[2:]

# Normalización y unión con el dataframe
for i in range(len(normal)):
    norm = pd.json_normalize(df[normal[i]])
    print(norm.shape)
    df.drop(normal[i], axis=1, inplace=True)
    df = df.join(norm)

# Revisión
df.head()
```

### Manejo de inconsistencias

```python
# Reemplazo de datos inconsistentes con la moda
print(df["Churn"].mode())
df["Churn"] = df["Churn"].apply(lambda x: "No" if (x != "Yes") & (x != "No") else x)
df["Churn"].value_counts()
```

```python
# Convertir columna categórica a numérica
df["Charges.Total"] = df["Charges.Total"].str.strip().replace("", "0")
df["Charges.Total"] = df["Charges.Total"].astype(np.float64)
df["Charges.Total"].head()
```

```python
df[["tenure", "Charges.Monthly", "Charges.Total"]].describe()
```

```python
num_col = ["tenure", "Charges.Monthly", "Charges.Total"]

print("Kurtosis de columnas numéricas importantes el dataset (para identificar atípicos)")

for i in num_col:
    print(f"Columna {i}: {df[i].kurtosis()}")
```

```python
fig, ax = plt.subplots(3, 1, figsize=(10, 10))
fig.suptitle("Distribución e identificación de datos atípicos en las columnas meses de contrato, carga mensual y carga total")
for i in range(len(num_col)):
    ax[i].boxplot(df[num_col[i]])
    ax[i].set_title(num_col[i])
    ax[i].xaxis.set_visible(False)
```

### Distribución de evasión

```python
import os

if not os.path.exists('img'):
    os.makedirs('img')

plt.figure(figsize=(12, 10))
plt.bar(df["Churn"].value_counts().index, df["Churn"].value_counts().values)
plt.title("Distribución de clientes que abandonaron")
plt.xlabel("Abandonaron (0) o no (1)")
plt.ylabel("Cantidad de clientes")
plt.savefig("img/Distribucion_de_clientes_que_abandonaron.png")
plt.show()
```

### Recuento de evasión por variables categóricas

```python
bar_cols = ["gender", "Contract", "PaymentMethod", "PhoneService"]

fig, ax = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Evasión de clientes por variables categóricas")
axes = ax.flatten()

for i, col in enumerate(bar_cols):
    cross = pd.crosstab(df[col], df["Churn"])
    cross.plot(kind="bar", ax=axes[i], rot=45, color=['red', 'green'])
    axes[i].set_title(f"Evasión de clientes por {bar_cols[i]}")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("")
    axes[i].set_ylim(0, 5000)
    axes[i].legend(title="Churn", labels=["No", "Sí"])

plt.savefig("img/Evasion_de_clientes_por_variables_categoricas.png")
plt.tight_layout(rect=[0, 0.03, 1, 1])
plt.show()
```

### Conteo de evasión por variables numéricas

```python
vio_cols = ["Charges.Total", "tenure", "Cuentas_Diarias"]

fig, ax = plt.subplots(3, 1, figsize=(12, 10))
fig.suptitle("Evasión de clientes por variables numéricas")
axes = ax.flatten()

for i in range(len(vio_cols)):
    sns.violinplot(data=df, x="Churn", y=vio_cols[i], ax=axes[i], color="green")
    axes[i].set_title(f"Evasión de clientes por {vio_cols[i]}")

plt.savefig("img/Evasion_de_clientes_por_variables_numericas.png")
plt.tight_layout()
plt.show()
```

### Limpieza y tratamiento de datos
Para este análisis, seguí una serie de pasos para preparar los datos:

Importación: Primero, extraje los datos de Github y los importé directamente como un DataFrame de la librería pandas en Python.

Limpieza: Los datos inicialmente eran bastante incómodos de leer. Por eso, realicé un proceso de normalización para que fueran mucho más legibles.

Procesamiento: Después de la limpieza, me enfoqué en las columnas más relevantes. Hice algo similar con la columna "Charges.Total", que también presentaba registros vacíos. ### Análisis Exploratorio de datos (EDA)
El 74% de los clientes está abandonando la compañía. He identificado algunas razones clave que podrían explicar este fenómeno:

Cancelación de contrato recurrente: Muchos clientes cancelan contratos que, por su extensión (uno o dos años), simplemente no pueden mantener. Solo un pequeño porcentaje de los clientes que permanecen tiene este tipo de contratos.

Cancelación por el método de pago: Los métodos de pago también parecen ser un factor. Algunos pueden resultar incómodos (como las facturas por correo), o podría haber problemas con los métodos automáticos, que, irónicamente, son los más convenientes. De hecho, la mayoría de los clientes que permanecen usan facturas electrónicas.

Servicios adicionales: Los clientes que se van no están aprovechando los servicios adicionales que ofrece la empresa, más allá del servicio telefónico. Hay muchos beneficios que la mayoría de los clientes que abandonan no adquieren, a diferencia de los que se quedan, cuyas cuentas diarias muestran gastos mayores.

Conclusiones e Insights
En resumen, la empresa está sufriendo un gran abandono de clientes debido a contratos demasiado extendidos que los clientes aceptan, posibles problemas con los métodos de pago y el escaso uso de los servicios adicionales que se ofrecen.

Los clientes que abandonaron la compañía son los que más se ajustan a estas características. La mayoría de estas conclusiones se basan en el comportamiento de los clientes que aún permanecen, lo que nos da pistas valiosas sobre cómo retener tanto a los clientes actuales como a los futuros.



### Recomendaciones
Aquí mis sugerencias para revertir esta situación:

Visibilidad de los servicios adicionales: Sugiero crear publicidad o días de prueba para los diferentes servicios adicionales. Esto aumentaría su visibilidad, ya sea a través de anuncios en redes sociales o con límites de días de prueba (por ejemplo, 10, 15 o 30 días).

Investigar métodos de pago: Es crucial investigar a fondo los métodos de pago automáticos, como las transferencias bancarias o las tarjetas de crédito. Podría haber problemas subyacentes como tiempos de facturación tardíos, aumentos de precios sospechosos o cualquier otra cuestión de seguridad que esté afectando la confianza del cliente.

Reducción de tiempo de contrato: Reducir la duración de los contratos podría ayudar a establecer compromisos más cortos y manejables con los clientes, ofreciendo opciones como contratos trimestrales, cuatrimestrales o semestrales.
